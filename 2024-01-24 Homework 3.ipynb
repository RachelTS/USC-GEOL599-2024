```python
import pandas as pd
import seaborn as sns
```

```python
# read the file
file= "C:/Users/rache/Documents/University of Southern California/2023-2024 Year 4/GEOL 599 Environmental data science/data/HW3_LA_AQS_2023.csv"
df= pd.read_csv(file)
df.head()
```

```python
# print out unique variables in Parameter Name and Duration Description columns
print(df["Parameter Name"].unique()) # "Ozone"
print(df["Duration Description"].unique()) # "1 HOUR"
```

```python
# Subset ozone hourly data
df_O3= df[(df["Parameter Name"]== "Ozone") & (df["Duration Description"]== "1 HOUR")]
df_O3.head()
```

```python
# Convert the dtype from object to date
df_O3["Date (Local)"] = pd.to_datetime(df_O3["Date (Local)"])
```

```python
# Convert ozone from ppm to ppb
df_O3["Arithmetic Mean"] = df_O3["Arithmetic Mean"]*1000
```

```python
# # See what it looks like using matplotlib.pyplot
# import matplotlib.pyplot as plt
# fig, ax = plt.subplots(figsize= (12,6)) #Changes the figure size that is spitted out
# ax.plot(date,value)
```

```python
# Plot data with Seaborn
fig = sns.relplot(data= df_O3, x= "Date (Local)", y= "Arithmetic Mean", kind= "line", height= 6, aspect= 1.5)
fig.set(xlabel ="month in 2023", ylabel = "ozone (ppbv)")
```

```python
# Find the highest value of O3 in this data subset
df_O3["Arithmetic Mean"].max() # Max is 45.75 ppbv
max_O3 = df_O3[df_O3["Arithmetic Mean"] == 45.75]
print(max_O3["Date (Local)"]) 
### Day of highest ozone is 2023-05-14
```

```python
# Find the number of days O3 exceeded 35 ppbv
df_O3_35plus = df_O3[df_O3["Arithmetic Mean"]>35]
len(df_O3_35plus) 
### There are 43 days where ozone is above 35 ppbv.
```
